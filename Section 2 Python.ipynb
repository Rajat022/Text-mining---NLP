```python
import os
import pandas as pd
import nltk
import operator
from collections import defaultdict
from itertools import combinations
from collections import defaultdict
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

# Loading the data
train_folder_path = "C:/Users/CYTech Student/OneDrive/Desktop/Reuters_Dataset/Reuters_Dataset/training"
test_folder_path = "C:/Users/CYTech Student/OneDrive/Desktop/Reuters_Dataset/Reuters_Dataset/test"

# Read the train files and extract the data
train_file_paths = [os.path.join(root, filename) for root, dirs, files in os.walk(train_folder_path) for filename in files]
train_texts = []
train_subjects = []

for file_path in train_file_paths:
    with open(file_path, "r") as file:
        train_texts.append(file.read())
    train_subjects.append(os.path.basename(os.path.dirname(file_path)))

train_df = pd.DataFrame({"subject": train_subjects, "content": train_texts})

# Read the test files and extract the data
test_file_paths = [os.path.join(root, filename) for root, dirs, files in os.walk(test_folder_path) for filename in files]
test_texts = []
test_subjects = []

for file_path in test_file_paths:
    with open(file_path, "r") as file:
        test_texts.append(file.read())
    test_subjects.append(os.path.basename(os.path.dirname(file_path)))

test_df = pd.DataFrame({"subject": test_subjects, "content": test_texts})

# Define stopwords
stop_words = set(stopwords.words('english'))

# Define lemmatizer
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # Convert to lowercase and remove punctuation
    text = text.lower()
    text = ''.join([c for c in text if c.isalpha() or c.isspace()])

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stopwords and lemmatize tokens
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]

    return tokens

# Tokenize and preprocess the content in train dataframe
train_df['tokens'] = train_df['content'].apply(preprocess_text)

# Tokenize and preprocess the content in test dataframe
test_df['tokens'] = test_df['content'].apply(preprocess_text)

# Dictionary to store the number of documents and number of occurrences of words
document_frequencies = defaultdict(int)
word_frequencies = defaultdict(int)

# Preprocess in train dataframe and test dataframe 
for tokens in train_df['tokens']:
    unique_tokens = set(tokens)
    for token in unique_tokens:
        document_frequencies[token] += 1
    for token in tokens:
        word_frequencies[token] += 1

for tokens in test_df['tokens']:
    unique_tokens = set(tokens)
    for token in unique_tokens:
        document_frequencies[token] += 1
    for token in tokens:
        word_frequencies[token] += 1

# Sort the document frequencies dictionary by value in descending order
sorted_frequencies = sorted(document_frequencies.items(), key=lambda x: x[1], reverse=True)

# Print the number of documents and number of occurrences for each word in descending order
for word, doc_freq in sorted_frequencies:
    word_freq = word_frequencies[word]
    print("Word:", word)
    print("Number of Occurrences:", word_freq)
    print("Number of Documents:", doc_freq)
    print()  # Print a blank line for readability

min_word_freq = 2  # Minimum number of occurrences for a termset to be considered frequent
frequent_termsets = [termset for termset, freq in word_frequencies.items() if freq >= min_word_freq]    
cluster_candidates = defaultdict(list)

for termset in frequent_termsets:
    for index, tokens in enumerate(train_df['tokens']):
        if set(termset).issubset(tokens):
            cluster_candidates[termset].append(index)           

for termset in frequent_termsets:
    print(termset)

# Cluster candidates for each frequent termset
cluster_candidates_new = defaultdict(set)
for frequent_termset in frequent_termsets:
    for index, tokens in enumerate(train_df['tokens']):
        if all(term in tokens for term in frequent_termset):
            folder_name = train_subjects[index] 
            cluster_candidates_new[frequent_termset].add(folder_name)
```
